```python
df = pd.read_csv('house_cleaned.csv')
df.head()
```

```python
cat_var = df.select_dtypes(include='object')
cat_var.head()
```

```python
miss_var_pe = cat_var.isnull().sum()/cat_var.shape[0]*100
miss_var_pe
```

```python
miss_var_per = cat_var.isnull().mean()*100
miss_var_per
```

```python
drop_var = miss_var_pe[miss_var_pe>20].keys()
```

```python
drop_var
```

```python
cat_var.drop(columns=drop_var, inplace=True)
cat_var.shape
```

```python
isnull_per = cat_var.isnull().mean()*100
miss_var = isnull_per[isnull_per>0].keys()
miss_var
```

```python
cat_var['address'].mode()
```

```python
cat_var['address'].value_counts()
```

```python
cat_var['address'].fillna(cat_var['address'].mode()[0])
```

```python
for var in miss_var:
  cat_var[var].fillna(cat_var[var].mode()[0], inplace=True)
  print(var, "=", cat_var[var].mode()[0])
```

```python
cat_var.isnull().sum()
```

```python
plt.figure(figsize = (15, 10))
for i, var in enumerate(miss_var):
    plt.subplot(34, 3, i+1)
    plt.hist(cat_var[var], label = "imput")
    plt.hist(df[var].dropna(), label = "Original")
    plt.legend()
```

```python
df.update(cat_var)
df.drop(columns=drop_var, inplace=True)
```
